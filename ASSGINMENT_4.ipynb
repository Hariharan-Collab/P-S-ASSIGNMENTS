```python
from google.colab import drive
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

drive.mount('/content/drive')
file_path = '/content/car_data_250 (1).csv'
df = pd.read_csv(file_path)

plt.figure(figsize=(10,6))
plt.hist(df['Price'], bins=20, edgecolor='black', alpha=0.7)
plt.xlabel('Price (USD)')
plt.ylabel('Number of Cars')
plt.title('Distribution of Car Prices')
plt.show()

common_price_range = df['Price'].value_counts().idxmax()
print("Most common price range:", common_price_range)

plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Horsepower', y='Mileage', hue='Fuel Type', palette='Set1', alpha=0.7)
z = np.polyfit(df['Horsepower'], df['Mileage'], 1)
p = np.poly1d(z)
plt.plot(df['Horsepower'], p(df['Horsepower']), color='black', linestyle='dashed')
plt.xlabel('Horsepower')
plt.ylabel('Mileage (MPG)')
plt.title('Horsepower vs. Mileage')
plt.show()

plt.figure(figsize=(12,6))
df['Brand'].value_counts().plot(kind='bar', color='skyblue')
plt.xlabel('Brand')
plt.ylabel('Number of Cars')
plt.title('Number of Cars by Brand')
plt.xticks(rotation=45)
plt.show()

most_popular_brand = df['Brand'].value_counts().idxmax()
print("Most popular car brand:", most_popular_brand)

df_grouped = df.groupby(['Year', 'Fuel Type']).size().unstack()
df_grouped.plot(kind='bar', figsize=(14,6))
plt.xlabel('Manufacturing Year')
plt.ylabel('Number of Cars')
plt.title('Fuel Type Preference Over the Years')
plt.legend(title='Fuel Type')
plt.show()
```
