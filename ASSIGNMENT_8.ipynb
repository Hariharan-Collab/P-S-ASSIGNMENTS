```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr

data = {
    'Daily Calorie Intake': [2000, 2200, 2500, 1800, 2300, 2600, 2100, 1900, 2700, 3000],
    'Exercise Duration': [60, 50, 30, 90, 45, 20, 75, 95, 15, 10],
    'BMI': [25, 26, 28, 23, 27, 30, 24, 22, 31, 33]
}

df = pd.DataFrame(data)

# Question 1: Create scatter plots
plt.figure(figsize=(10, 6))
plt.subplot(1, 3, 1)
sns.scatterplot(x='Daily Calorie Intake', y='BMI', data=df)
plt.title('Daily Calorie Intake vs. BMI')

plt.subplot(1, 3, 2)
sns.scatterplot(x='Exercise Duration', y='BMI', data=df)
plt.title('Exercise Duration vs. BMI')

plt.subplot(1, 3, 3)
sns.scatterplot(x='Daily Calorie Intake', y='Exercise Duration', data=df)
plt.title('Daily Calorie Intake vs. Exercise Duration')

plt.tight_layout()
plt.show()

# Question 2: Calculate Pearson correlation coefficients
corr_calories_bmi, _ = pearsonr(df['Daily Calorie Intake'], df['BMI'])
corr_exercise_bmi, _ = pearsonr(df['Exercise Duration'], df['BMI'])
corr_calories_exercise, _ = pearsonr(df['Daily Calorie Intake'], df['Exercise Duration'])

print(f"Pearson correlation coefficient (Daily Calorie Intake vs. BMI): {corr_calories_bmi:.2f}")
print(f"Pearson correlation coefficient (Exercise Duration vs. BMI): {corr_exercise_bmi:.2f}")
print(f"Pearson correlation coefficient (Daily Calorie Intake vs. Exercise Duration): {corr_calories_exercise:.2f}")

# Question 3: Interpret the results
print("\nInterpretation:")
if corr_calories_bmi > 0:
    print("There is a positive correlation between Daily Calorie Intake and BMI, indicating that as calorie intake increases, BMI tends to increase.")
else:
    print("There is a negative correlation between Daily Calorie Intake and BMI, indicating that as calorie intake increases, BMI tends to decrease.")

if corr_exercise_bmi < 0:
    print("There is a negative correlation between Exercise Duration and BMI, suggesting that longer exercise durations are associated with lower BMI.")
else:
    print("There is a positive correlation between Exercise Duration and BMI, suggesting that more exercise is associated with a higher BMI.")

if corr_calories_exercise > 0:
    print("There is a positive correlation between Daily Calorie Intake and Exercise Duration, indicating that higher calorie intake tends to be associated with more exercise.")
else:
    print("There is a negative correlation between Daily Calorie Intake and Exercise Duration, suggesting that higher calorie intake may be linked with less exercise.")
```
